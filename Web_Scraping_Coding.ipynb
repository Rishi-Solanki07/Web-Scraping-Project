```python

response = requests.get('https://www.netflix.com/tudum/top10/')
response.raise_for_status()
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup)
```

```python
movie_details = soup.find('tbody').find_all('tr')
print(len(movie_details))
```

```python
movie_details = soup.find('tbody').find_all('tr')

for movie_detail in movie_details:
    # Find the elements inside each <tr> tag
    rank_element = movie_detail.find('td', class_="pb-2 tbl-cell tbl-cell-rank")
    name_element = movie_detail.find('td', class_="pb-2 tbl-cell tbl-cell-name")
    weeks_element = movie_detail.find('td', class_="pb-2 tbl-cell tbl-cell-weeks")
    hours_element = movie_detail.find('td', class_="pb-2 tbl-cell tbl-cell-hours")
    runtime_element = movie_detail.find('td', class_="pb-2 tbl-cell tbl-cell-runtime")
    views_element = movie_detail.find('td', class_="pb-2 tbl-cell tbl-cell-vhor")
    
    # Get the text from each element or set a default value if element is not found
    rank = rank_element.text if rank_element else "N/A"
    name = name_element.text if name_element else "N/A"
    weeks_in_top_10 = weeks_element.text if weeks_element else "N/A"
    hours_viewed = hours_element.text if hours_element else "N/A"
    run_time = runtime_element.text if runtime_element else "N/A"
    views = views_element.text if views_element else "N/A"
    
    # Print the details of each movie
    print(rank, name, weeks_in_top_10, hours_viewed, run_time, views)
```

```python
Top_10_Movies = []

# first block of code
movie_details = soup.find('tbody').find_all('tr')
for movie_detail in movie_details:
    # Find the elements inside each <tr> tag
    rank_element = movie_detail.find('td', class_="pb-2 tbl-cell tbl-cell-rank pt-2")
    name_element = movie_detail.find('td', class_="pb-2 tbl-cell tbl-cell-name pt-2")
    weeks_element = movie_detail.find('td', class_="pb-2 tbl-cell tbl-cell-weeks pt-2")
    hours_element = movie_detail.find('td', class_="pb-2 tbl-cell tbl-cell-hours pt-2")
    runtime_element = movie_detail.find('td', class_="pb-2 tbl-cell tbl-cell-runtime pt-2")
    views_element = movie_detail.find('td', class_="pb-2 tbl-cell tbl-cell-vhor pt-2")
    
    # Get the text from each element or set a default value if element is not found
    rank = rank_element.text if rank_element else "N/A"
    name = name_element.text if name_element else "N/A"
    weeks_in_top_10 = weeks_element.text if weeks_element else "N/A"
    hours_viewed = hours_element.text if hours_element else "N/A"
    run_time = runtime_element.text if runtime_element else "N/A"
    views = views_element.text if views_element else "N/A"
    
    # Append the details of each movie to the list
    Top_10_Movies.append((rank, name, weeks_in_top_10, hours_viewed, run_time, views))
    
    
 # second block of code
movie_details = soup.find('tbody').find_all('tr')
for movie_detail in movie_details:
    # Find the elements inside each <tr> tag
    rank_element = movie_detail.find('td', class_="pb-2 tbl-cell tbl-cell-rank")
    name_element = movie_detail.find('td', class_="pb-2 tbl-cell tbl-cell-name")
    weeks_element = movie_detail.find('td', class_="pb-2 tbl-cell tbl-cell-weeks")
    hours_element = movie_detail.find('td', class_="pb-2 tbl-cell tbl-cell-hours")
    runtime_element = movie_detail.find('td', class_="pb-2 tbl-cell tbl-cell-runtime")
    views_element = movie_detail.find('td', class_="pb-2 tbl-cell tbl-cell-vhor")
    
    # Get the text from each element or set a default value if element is not found
    rank = rank_element.text if rank_element else "N/A"
    name = name_element.text if name_element else "N/A"
    weeks_in_top_10 = weeks_element.text if weeks_element else "N/A"
    hours_viewed = hours_element.text if hours_element else "N/A"
    run_time = runtime_element.text if runtime_element else "N/A"
    views = views_element.text if views_element else "N/A"
    
    # Append the details of each movie to the list
    Top_10_Movies.append((rank, name, weeks_in_top_10, hours_viewed, run_time, views))
    
# Remove None values from the list
Top_10_Movies = [movie for movie in Top_10_Movies if 'N/A' not in movie]

# Print the merged results
for movie in Top_10_Movies:
    print(movie)    
```
